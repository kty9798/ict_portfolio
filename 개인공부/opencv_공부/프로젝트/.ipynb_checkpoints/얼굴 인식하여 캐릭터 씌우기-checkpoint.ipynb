```python
import cv2
import mediapipe as mp

#얼굴을 찾고 찾은 얼굴에 표시를 해주기 위한 변수
mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils #
# 얼굴의 특징을 그리기 위한 drawing units 모듈을 사용


cap = cv2.VideoCapture("face.mp4")
#클로즈 할 필요 없이                             #0 or 1 

#이미지 파일 불러오기
image_right_eye = cv2.imread("right_eye.png") #100 100
image_left_eye = cv2.imread("left_eye.png") #100 100 
image_nose = cv2.imread("nose_eye.png") #300 100

with mp_face_detection.FaceDetection(model_selection=0, min_detection_confidence=0.7) as face_detection:
    while cap.isOpened():
        success, image = cap.read() #frame을 가져옴
        if not success:
            break
      
        # To improve performance, optionally mark the image as not writeable to
        # pass by reference.
        image.flags.writeable = False
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) #rgb로
        results = face_detection.process(image)
        #rgb 기준으로 처리하는듯
        # Draw the face detection annotations on the image.
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        
        if results.detections: #결과가 있으면
            #6개의 특징 오른눈, 왼눈 ,코 끝부분 ,입중심 ,오른쪽 귀 ,왼쪽 귀(귀 구슬점)
            for detection in results.detections:
                #mp_drawing.draw_detection(image, detection) #점을 찍는것
                print(detection)
                #그려주는 것 
                
                #특정 위치 
                key_point= detection.location_data.relative_keypoints
                right_eye = key_point[0]
                left_eye = key_point[1]
                nose_tip = key_point[2]
                #이미지 비율로 적용
                h , w , _ = image.shape #높이 너비 채널
                
                right_eye = (int(right_eye.x * w) , int(right_eye.y * h)) #실제 좌표 
                #튜플 형태로
                left_eye = (int(left_eye.x * w) , int(left_eye.y* h))
                nose_tip = (int(nose_tip.x * w) , int(nose_tip.y* h))
                #양 눈에 동그라미 그리기
                #cv2.circle(image , right_eye , 50 , (255, 0 , 0) , 10 , cv2.LINE_AA)#파란색
                
                #cv2.circle(image , left_eye , 50 , (0, 0 , 255) , 10 , cv2.LINE_AA)#파란색
                #cv2.circle(image , nose_tip , 75 , (0, 255 , 0) , 10 , cv2.LINE_AA)#파란색
                
                
                image[right_eye[1]-50: right_eye[1]+50]
                
                
                
                
                #이미지 씌우게 되면
        
        # Flip the image horizontally for a selfie-view display.
        cv2.imshow('MediaPipe Face Detection',cv2.resize(image , None , fx = 0.5 , fy = 0.5 ))
        
        
        if cv2.waitKey(4) == ord("q"):
              break

cap.release()
cv2.destroyAllWindows()
```
