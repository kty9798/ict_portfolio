## 18. 이미지 검출(윤곽선)

### 윤곽선(Contour) :경계선을 연결한 선

```python
import cv2
import numpy as np
img = cv2.imread("poker.png")
#원본 이미지

target_img = img.copy()

#binary 처리해야됨

gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)

ret , otsu = cv2.threshold(gray , -1 , 255 , cv2.THRESH_BINARY | cv2.THRESH_OTSU)
#오츠 알고리즘
                                            #윤곽선 찾는 모드
contours  , hierachy = cv2.findContours(otsu , cv2.RETR_LIST , cv2.CHAIN_APPROX_NONE)
#윤곽선 정보 , 구조
#이미지 윤곽선 찾는 모드 (mode) , 윤곽선 찾을떄 사용하는 근사치 방법

COLOR = (0 , 200 ,0) #녹색
cv2.drawContours(target_img , contours , -1 , COLOR , 2) #모든 윤곽선 그려라
#대상이미지 , 윤곽선 정보 , 인덱스 , 색깔 , 두께




cv2.imshow('img'  , img)
cv2.imshow('gray'  , gray)
cv2.imshow('target'  , target_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 윤곽선 찾기 모드
1. cv2.RETR_EXTERNAL:가장 외곽의 윤곽선만 찾음
2. cv2.RETR_LIST: 모든 윤곽선 찾음(계층정보 없음)
3. cv2.RETR_TREE: 모든 윤곽선 찾음(계층 정보를 트리구조로 생성)

```python
import cv2
import numpy as np
img = cv2.imread("poker.png")
#원본 이미지

target_img = img.copy()

#binary 처리해야됨

gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)

ret , otsu = cv2.threshold(gray , -1 , 255 , cv2.THRESH_BINARY | cv2.THRESH_OTSU)
#오츠 알고리즘
                                            #윤곽선 찾는 모드
#contours  , hierachy = cv2.findContours(otsu , cv2.RETR_LIST , cv2.CHAIN_APPROX_NONE)
contours  , hierachy = cv2.findContours(otsu , cv2.RETR_TREE , cv2.CHAIN_APPROX_NONE)
#윤곽선 정보 , 구조
#이미지 윤곽선 찾는 모드 (mode) , 윤곽선 찾을떄 사용하는 근사치 방법 : CHAIN_APPROX_NONE , CHAIN_APPROX_SIMPLE #사각형이면 4점만

print(hierachy)
COLOR = (0 , 200 ,0) #녹색
cv2.drawContours(target_img , contours , -1 , COLOR , 2) #모든 윤곽선 그려라
#대상이미지 , 윤곽선 정보 , 인덱스 , 색깔 , 두께




cv2.imshow('img'  , img)
cv2.imshow('gray'  , gray)
cv2.imshow('target'  , target_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 경계 사각형
윤곽선의 경계면을 둘러싸는 사각형


```python
import cv2
import numpy as np
img = cv2.imread("poker.png")
#원본 이미지

target_img = img.copy()

#binary 처리해야됨

gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)

ret , otsu = cv2.threshold(gray , -1 , 255 , cv2.THRESH_BINARY | cv2.THRESH_OTSU)


contours  , hierachy = cv2.findContours(otsu , cv2.RETR_LIST , cv2.CHAIN_APPROX_NONE)



COLOR = (0 , 200 ,0) #녹색
#cv2.drawContours(target_img , contours , -1 , COLOR , 2) #모든 윤곽선 그려라

for cnt in contours:
    x , y , width , height = cv2.boundingRect(cnt)
    cv2.rectangle(target_img ,  (x , y) , (x+width , y+ height) , COLOR , 2) #사각형 그림
    




cv2.imshow('img'  , img)
cv2.imshow('gray'  , gray)
cv2.imshow('target'  , target_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 미니 프로젝트 : 개별 카드 추출해서 파일 저장

```python
import cv2
import numpy as np
img = cv2.imread("poker.png")
#원본 이미지

target_img = img.copy()

#binary 처리해야됨

gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)

ret , otsu = cv2.threshold(gray , -1 , 255 , cv2.THRESH_BINARY | cv2.THRESH_OTSU)


contours  , hierachy = cv2.findContours(otsu , cv2.RETR_LIST , cv2.CHAIN_APPROX_NONE)



COLOR = (0 , 200 ,0) #녹색
#cv2.drawContours(target_img , contours , -1 , COLOR , 2) #모든 윤곽선 그려라

idx = 1
for cnt in contours:
    if cv2.contourArea(cnt) > 25000:
        x , y , width , height = cv2.boundingRect(cnt)
        cv2.rectangle(target_img ,  (x , y) , (x+width , y+ height) , COLOR , 2) #사각형 그림
        
        crop = img[y : y+height , x : x+width]
        cv2.imshow("crop_{}".format(idx) , crop)
        cv2.imwrite(f"card_crop_{idx}.png" , crop) #파일 저장
        idx +=1
    


cv2.imshow('target'  , target_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
